```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

def calculate_weekly_returns(df):
    weekly_data = df.resample('W').last()
    weekly_returns = weekly_data.pct_change().dropna()
    return weekly_returns

def rank_stocks(weekly_returns, window=50):
    avg_returns = weekly_returns.rolling(window=window).mean()
    latest_avg_returns = avg_returns.iloc[-1].sort_values(ascending=False)
    return latest_avg_returns.index.tolist()

def get_top_stocks(ranked_stocks, n=20):
    return ranked_stocks[:n]

def portfolio_objective(weights, cov_matrix, prev_weights, lambda_cost):
    # Calculate variance (risk) component
    variance = np.dot(weights.T, np.dot(cov_matrix, weights))

    # Calculate transaction cost component
    if prev_weights is not None:
        transaction_costs = lambda_cost * np.sum(np.abs(weights - prev_weights))
    else:
        transaction_costs = 0

    return variance + transaction_costs

def create_min_variance_portfolio(returns, prev_weights=None, lambda_cost=0.0050):
    """
    Create a minimum variance portfolio with transaction costs

    Parameters:
    returns: DataFrame of asset returns
    prev_weights: Previous portfolio weights (if None, assumes no previous positions)
    lambda_cost: Transaction cost parameter (50 basis points = 0.0050)
    """
    n = len(returns.columns)

    # Calculate covariance matrix
    cov_matrix = returns.cov().values

    # If no previous weights provided, initialize with zeros
    if prev_weights is None:
        prev_weights = np.zeros(n)

    # Define constraints
    constraints = [
        {'type': 'eq', 'fun': lambda x: np.sum(x) - 1}  # weights sum to 1
    ]

    # Define bounds (0 to 1 for long-only portfolio)
    bounds = tuple((0, 1) for _ in range(n))

    # Initial guess (use previous weights if available)
    x0 = prev_weights if prev_weights is not None else np.array([1/n] * n)

    # Optimize
    result = minimize(
        portfolio_objective,
        x0,
        args=(cov_matrix, prev_weights, lambda_cost),
        method='SLSQP',
        bounds=bounds,
        constraints=constraints,
        options={'ftol': 1e-9, 'maxiter': 1000}
    )

    if not result.success:
        print(f"Warning: Optimization failed: {result.message}")

    return result.x

def get_stock_weightage(weekly_returns, current_date, prev_weights=None):
    data_slice = weekly_returns.loc[:current_date].iloc[-50:]

    if len(data_slice) < 50:
        raise ValueError("Not enough historical data before the target date.")

    ranked_stocks = rank_stocks(data_slice)
    top_stocks = get_top_stocks(ranked_stocks)
    top_stock_returns = data_slice[top_stocks]

    # Align previous weights with current stock selection
    if prev_weights is not None:
        aligned_prev_weights = np.zeros(len(top_stocks))
        for i, stock in enumerate(top_stocks):
            if stock in prev_weights:
                aligned_prev_weights[i] = prev_weights[stock]
        prev_weights = aligned_prev_weights

    # Get optimal weights
    weights = create_min_variance_portfolio(
        top_stock_returns,
        prev_weights,
        lambda_cost=0.0050  # 50 basis points transaction cost
    )

    # Create dictionary of stock weights
    stock_weights = dict(zip(top_stocks, weights))

    # Ensure weights sum to 1 (handle numerical precision issues)
    weight_sum = sum(stock_weights.values())
    if abs(weight_sum - 1.0) > 1e-5:
        for stock in stock_weights:
            stock_weights[stock] /= weight_sum

    return stock_weights

def calculate_drawdown(cumulative_returns):
    drawdown = cumulative_returns / cumulative_returns.cummax() - 1
    max_drawdown = drawdown.min()
    return drawdown, max_drawdown

def sortino_ratio(returns, target=0, periods_per_year=52):
    downside_returns = returns[returns < target]
    downside_deviation = np.sqrt((downside_returns ** 2).mean()) * np.sqrt(periods_per_year)
    annualized_return = returns.mean() * periods_per_year
    return annualized_return / downside_deviation if downside_deviation != 0 else np.nan

def calmar_ratio(annualized_return, max_drawdown):
    return annualized_return / abs(max_drawdown) if max_drawdown != 0 else np.nan

def backtest_strategy(df):
    weekly_returns = calculate_weekly_returns(df)
    portfolio_returns = []
    weightage_history = []
    prev_weights = None

    for i in range(50, len(weekly_returns)):
        current_date = weekly_returns.index[i-1]
        next_week = weekly_returns.index[i]

        try:
            weightage = get_stock_weightage(weekly_returns.iloc[:i], current_date, prev_weights)
            top_stocks = list(weightage.keys())
            weights = list(weightage.values())

            # Calculate transaction costs
            if prev_weights is not None:
                transaction_costs = 0.0050 * sum(  # 50 basis points
                    abs(weightage.get(stock, 0) - prev_weights.get(stock, 0))
                    for stock in set(weightage.keys()) | set(prev_weights.keys())
                )
            else:
                transaction_costs = 0

            # Calculate returns after transaction costs
            next_week_return = weekly_returns.loc[next_week][top_stocks].dot(weights) - transaction_costs
            portfolio_returns.append(next_week_return)
            weightage_history.append((next_week, weightage))

            prev_weights = weightage

        except ValueError as e:
            print(f"Skipping week {next_week} due to error: {e}")

    valid_index = weekly_returns.index[50:50+len(portfolio_returns)]
    return pd.Series(portfolio_returns, index=valid_index), weightage_history

def analyze_and_plot_results(portfolio_returns, weightage_history):
    # Calculate cumulative returns and metrics
    cumulative_returns = (1 + portfolio_returns).cumprod()
    total_return = cumulative_returns.iloc[-1]
    profit_only_return = total_return - 1
    annualized_return = (total_return ** (52/len(cumulative_returns)) - 1)
    annualized_volatility = portfolio_returns.std() * np.sqrt(52)
    sharpe_ratio = portfolio_returns.mean() / portfolio_returns.std() * np.sqrt(52)

    # Calculate drawdown and additional metrics
    drawdown, max_drawdown = calculate_drawdown(cumulative_returns)
    calmar = calmar_ratio(annualized_return, max_drawdown)
    sortino = sortino_ratio(portfolio_returns)

    # Print backtest results
    print("\nBacktest Results:")
    print(f"Total Return (including initial capital): {total_return:.4f}")
    print(f"This means for every $1 invested, you now have ${total_return:.2f}")
    print(f"Profit-only Return: {profit_only_return:.4%}")
    print(f"Annualized Return: {annualized_return:.2%}")
    print(f"Annualized Volatility: {annualized_volatility:.2%}")
    print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
    print(f"Sortino Ratio: {sortino:.2f}")
    print(f"Max Drawdown: {max_drawdown:.2%}")
    print(f"Calmar Ratio: {calmar:.2f}")

    # Plot cumulative returns and drawdown
    plt.figure(figsize=(12, 8))

    plt.subplot(2, 1, 1)
    cumulative_returns.plot()
    plt.title('Cumulative Portfolio Value (Starting from $1)')
    plt.xlabel('Date')
    plt.ylabel('Portfolio Value')
    plt.grid(True)

    plt.subplot(2, 1, 2)
    drawdown.plot(color='red')
    plt.title('Drawdown Over Time')
    plt.xlabel('Date')
    plt.ylabel('Drawdown')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    return {
        'total_return': total_return,
        'annualized_return': annualized_return,
        'annualized_volatility': annualized_volatility,
        'sharpe_ratio': sharpe_ratio,
        'sortino_ratio': sortino,
        'max_drawdown': max_drawdown,
        'calmar_ratio': calmar
    }

def main():
    # Load the data
    print("Loading data...")
    df = pd.read_csv('total_df.csv', index_col='Date', parse_dates=True)

    # Run backtest
    print("\nRunning backtest...")
    portfolio_returns, weightage_history = backtest_strategy(df)

    # Analyze and plot results
    print("\nAnalyzing results...")
    metrics = analyze_and_plot_results(portfolio_returns, weightage_history)

    # Get weightage for the last date in the dataset
    last_date = df.index[-1]
    try:
        print("\nCalculating final portfolio weights...")
        weightage = get_stock_weightage(calculate_weekly_returns(df), last_date)
        print(f"\nStock weightage for week of {last_date.date()}:")
        for stock, weight in sorted(weightage.items(), key=lambda x: x[1], reverse=True):
            print(f"{stock}: {weight:.4f}")
    except ValueError as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

def calculate_weekly_returns(df):
    weekly_data = df.resample('W').last()
    weekly_returns = weekly_data.pct_change().dropna()
    return weekly_returns

def rank_stocks(weekly_returns, window=50):
    avg_returns = weekly_returns.rolling(window=window).mean()
    latest_avg_returns = avg_returns.iloc[-1].sort_values(ascending=False)
    return latest_avg_returns.index.tolist()

def get_top_stocks(ranked_stocks, n=20):
    return ranked_stocks[:n]

def marcenko_pastur_pdf(x, Q, sigma=1):
    """Calculate Marcenko-Pastur probability density at x"""
    y = 1/Q
    b = np.power(sigma * (1 + np.sqrt(1/Q)), 2)  # Upper bound
    a = np.power(sigma * (1 - np.sqrt(1/Q)), 2)  # Lower bound

    return (Q/(2*np.pi*sigma*sigma*x)) * np.sqrt((b-x)*(x-a)) if a <= x <= b else 0

def find_eigenvalue_cutoff(eigenvalues, N, T):
    """
    Find the eigenvalue cutoff point using Marcenko-Pastur distribution
    N: number of assets
    T: number of observations
    """
    Q = N/T
    sigma = np.std(eigenvalues)

    # Calculate theoretical distribution
    x = np.linspace(0, max(eigenvalues), 1000)
    mp_density = [marcenko_pastur_pdf(xi, Q, sigma) for xi in x]

    # Calculate empirical distribution
    kde = gaussian_kde(eigenvalues)
    empirical_density = kde(x)

    # Find intersection point
    diff = empirical_density - mp_density
    cutoff_idx = np.where(diff > 0)[0][0]
    return x[cutoff_idx]

def denoise_correlation_matrix(returns):
    """
    Implement the de-noising algorithm using eigenvalue adjustment
    """
    # Calculate sample covariance matrix
    cov_matrix = returns.cov().values
    correlation_matrix = returns.corr().values

    # Perform eigendecomposition
    eigenvalues, eigenvectors = np.linalg.eigh(correlation_matrix)
    eigenvalues = eigenvalues[::-1]
    eigenvectors = eigenvectors[:, ::-1]

    # Find cutoff eigenvalue
    N = len(returns.columns)
    T = len(returns)
    lambda_plus = find_eigenvalue_cutoff(eigenvalues, N, T)

    idx = np.where(eigenvalues <= lambda_plus)[0]
    if len(idx) == 0:
      raise ValueError("No eigenvalues are smaller than or equal to lambda_plus")
    i = idx[0]

    # Find position i where eigenvalues become noise
    i = np.where(eigenvalues <= lambda_plus)[0][0]

    # Adjust noisy eigenvalues
    adjusted_eigenvalues = eigenvalues.copy()
    for j in range(i, N):
        adjusted_eigenvalues[j] = np.mean(eigenvalues[i:])

    # Reconstruct correlation matrix
    Lambda_tilde = np.diag(adjusted_eigenvalues)
    C_tilde = eigenvectors @ Lambda_tilde @ eigenvectors.T

    # Rescale to ensure diagonal entries are 1
    scale = np.sqrt(np.diag(C_tilde))
    C = C_tilde / (scale[:, None] @ scale[None, :])

    # Convert back to covariance matrix
    std_dev = np.sqrt(np.diag(cov_matrix))
    denoised_cov = C * (std_dev[:, None] @ std_dev[None, :])

    return denoised_cov

def create_min_variance_portfolio(returns, prev_weights=None, lambda_cost=0.0050):
    """
    Create a minimum variance portfolio with transaction costs and de-noised covariance matrix
    """
    n = len(returns.columns)

    # Calculate de-noised covariance matrix
    cov_matrix = denoise_correlation_matrix(returns)

    # If no previous weights provided, initialize with zeros
    if prev_weights is None:
        prev_weights = np.zeros(n)

    def portfolio_objective(weights, cov_matrix, prev_weights, lambda_cost):
        variance = np.dot(weights.T, np.dot(cov_matrix, weights))
        if prev_weights is not None:
            transaction_costs = lambda_cost * np.sum(np.abs(weights - prev_weights))
        else:
            transaction_costs = 0
        return variance + transaction_costs

    constraints = [
        {'type': 'eq', 'fun': lambda x: np.sum(x) - 1}
    ]

    bounds = tuple((0, 1) for _ in range(n))
    x0 = prev_weights if prev_weights is not None else np.array([1/n] * n)

    result = minimize(
        portfolio_objective,
        x0,
        args=(cov_matrix, prev_weights, lambda_cost),
        method='SLSQP',
        bounds=bounds,
        constraints=constraints,
        options={'ftol': 1e-9, 'maxiter': 1000}
    )

    if not result.success:
        print(f"Warning: Optimization failed: {result.message}")

    return result.x

def get_stock_weightage(weekly_returns, current_date, prev_weights=None):
    data_slice = weekly_returns.loc[:current_date].iloc[-50:]

    if len(data_slice) < 50:
        raise ValueError("Not enough historical data before the target date.")

    ranked_stocks = rank_stocks(data_slice)
    top_stocks = get_top_stocks(ranked_stocks)
    top_stock_returns = data_slice[top_stocks]

    # Align previous weights with current stock selection
    if prev_weights is not None:
        aligned_prev_weights = np.zeros(len(top_stocks))
        for i, stock in enumerate(top_stocks):
            if stock in prev_weights:
                aligned_prev_weights[i] = prev_weights[stock]
        prev_weights = aligned_prev_weights

    # Get optimal weights
    weights = create_min_variance_portfolio(
        top_stock_returns,
        prev_weights,
        lambda_cost=0.0050  # 50 basis points transaction cost
    )

    # Create dictionary of stock weights
    stock_weights = dict(zip(top_stocks, weights))

    # Ensure weights sum to 1 (handle numerical precision issues)
    weight_sum = sum(stock_weights.values())
    if abs(weight_sum - 1.0) > 1e-5:
        for stock in stock_weights:
            stock_weights[stock] /= weight_sum

    # Debug information
    weight_values = list(stock_weights.values())
    print(f"\nDate: {current_date}")
    print(f"Weight standard deviation: {np.std(weight_values):.4f}")
    print(f"Max weight: {max(weight_values):.4f}")
    print(f"Min weight: {min(weight_values):.4f}")

    return stock_weights

def calculate_drawdown(cumulative_returns):
    drawdown = cumulative_returns / cumulative_returns.cummax() - 1
    max_drawdown = drawdown.min()
    return drawdown, max_drawdown

def sortino_ratio(returns, target=0, periods_per_year=52):
    downside_returns = returns[returns < target]
    downside_deviation = np.sqrt((downside_returns ** 2).mean()) * np.sqrt(periods_per_year)
    annualized_return = returns.mean() * periods_per_year
    return annualized_return / downside_deviation if downside_deviation != 0 else np.nan

def calmar_ratio(annualized_return, max_drawdown):
    return annualized_return / abs(max_drawdown) if max_drawdown != 0 else np.nan

def backtest_strategy(df):
    weekly_returns = calculate_weekly_returns(df)
    portfolio_returns = []
    weightage_history = []
    prev_weights = None

    for i in range(50, len(weekly_returns)):
        current_date = weekly_returns.index[i-1]
        next_week = weekly_returns.index[i]

        try:
            weightage = get_stock_weightage(weekly_returns.iloc[:i], current_date, prev_weights)
            top_stocks = list(weightage.keys())
            weights = list(weightage.values())

            # Calculate transaction costs
            if prev_weights is not None:
                transaction_costs = 0.0050 * sum(  # 50 basis points
                    abs(weightage.get(stock, 0) - prev_weights.get(stock, 0))
                    for stock in set(weightage.keys()) | set(prev_weights.keys())
                )
            else:
                transaction_costs = 0

            # Calculate returns after transaction costs
            next_week_return = weekly_returns.loc[next_week][top_stocks].dot(weights) - transaction_costs
            portfolio_returns.append(next_week_return)
            weightage_history.append((next_week, weightage))

            prev_weights = weightage

        except ValueError as e:
            print(f"Skipping week {next_week} due to error: {e}")

    valid_index = weekly_returns.index[50:50+len(portfolio_returns)]
    return pd.Series(portfolio_returns, index=valid_index), weightage_history

def analyze_and_plot_results(portfolio_returns, weightage_history):
    # Calculate cumulative returns and metrics
    cumulative_returns = (1 + portfolio_returns).cumprod()
    total_return = cumulative_returns.iloc[-1]
    profit_only_return = total_return - 1
    annualized_return = (total_return ** (52/len(cumulative_returns)) - 1)
    annualized_volatility = portfolio_returns.std() * np.sqrt(52)
    sharpe_ratio = portfolio_returns.mean() / portfolio_returns.std() * np.sqrt(52)

    # Calculate drawdown and additional metrics
    drawdown, max_drawdown = calculate_drawdown(cumulative_returns)
    calmar = calmar_ratio(annualized_return, max_drawdown)
    sortino = sortino_ratio(portfolio_returns)

    # Print backtest results
    print("\nBacktest Results:")
    print(f"Total Return (including initial capital): {total_return:.4f}")
    print(f"This means for every $1 invested, you now have ${total_return:.2f}")
    print(f"Profit-only Return: {profit_only_return:.4%}")
    print(f"Annualized Return: {annualized_return:.2%}")
    print(f"Annualized Volatility: {annualized_volatility:.2%}")
    print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
    print(f"Sortino Ratio: {sortino:.2f}")
    print(f"Max Drawdown: {max_drawdown:.2%}")
    print(f"Calmar Ratio: {calmar:.2f}")

    # Plot cumulative returns and drawdown
    plt.figure(figsize=(12, 8))

    plt.subplot(2, 1, 1)
    cumulative_returns.plot()
    plt.title('Cumulative Portfolio Value (Starting from $1)')
    plt.xlabel('Date')
    plt.ylabel('Portfolio Value')
    plt.grid(True)

    plt.subplot(2, 1, 2)
    drawdown.plot(color='red')
    plt.title('Drawdown Over Time')
    plt.xlabel('Date')
    plt.ylabel('Drawdown')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # Plot weight evolution
    dates = [x[0] for x in weightage_history]
    all_stocks = set()
    for _, weights in weightage_history:
        all_stocks.update(weights.keys())

    weight_df = pd.DataFrame(index=dates, columns=list(all_stocks), dtype=float)

    for date, weights in weightage_history:
        for stock in all_stocks:
            weight_df.loc[date, stock] = weights.get(stock, 0)

    plt.figure(figsize=(12, 6))
    weight_df.plot.area(stacked=True)
    plt.title('Portfolio Weights Evolution Over Time')
    plt.xlabel('Date')
    plt.ylabel('Weight')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    return {
        'total_return': total_return,
        'annualized_return': annualized_return,
        'annualized_volatility': annualized_volatility,
        'sharpe_ratio': sharpe_ratio,
        'sortino_ratio': sortino,
        'max_drawdown': max_drawdown,
        'calmar_ratio': calmar
    }

def main():
    # Load the data
    print("Loading data...")
    df = pd.read_csv('total_df.csv', index_col='Date', parse_dates=True)

    # Run backtest
    print("\nRunning backtest with de-noised covariance matrices...")
    portfolio_returns, weightage_history = backtest_strategy(df)

    # Analyze and plot results
    print("\nAnalyzing results...")
    metrics = analyze_and_plot_results(portfolio_returns, weightage_history)

    # Get weightage for the last date in the dataset
    last_date = df.index[-1]
    try:
        print("\nCalculating final portfolio weights...")
        weightage = get_stock_weightage(calculate_weekly_returns(df), last_date)
        print(f"\nStock weightage for week of {last_date.date()}:")
        for stock, weight in sorted(weightage.items(), key=lambda x: x[1], reverse=True):
            print(f"{stock}: {weight:.4f}")
    except ValueError as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

def calculate_weekly_returns(df):
    weekly_data = df.resample('W').last()
    weekly_returns = weekly_data.pct_change().dropna()
    return weekly_returns

def rank_stocks(weekly_returns, window=50):
    avg_returns = weekly_returns.rolling(window=window).mean()
    latest_avg_returns = avg_returns.iloc[-1].sort_values(ascending=False)
    return latest_avg_returns.index.tolist()

def get_top_stocks(ranked_stocks, n=20):
    return ranked_stocks[:n]

def marcenko_pastur_pdf(x, Q, sigma=1):
    """Calculate Marcenko-Pastur probability density at x"""
    y = 1/Q
    b = np.power(sigma * (1 + np.sqrt(1/Q)), 2)  # Upper bound
    a = np.power(sigma * (1 - np.sqrt(1/Q)), 2)  # Lower bound

    return (Q/(2*np.pi*sigma*sigma*x)) * np.sqrt((b-x)*(x-a)) if a <= x <= b else 0

def find_eigenvalue_cutoff(eigenvalues, N, T):
    """
    Find the eigenvalue cutoff point using Marcenko-Pastur distribution
    N: number of assets
    T: number of observations
    """
    Q = N/T
    sigma = np.std(eigenvalues)

    # Calculate theoretical distribution
    x = np.linspace(0, max(eigenvalues), 1000)
    mp_density = [marcenko_pastur_pdf(xi, Q, sigma) for xi in x]

    # Calculate empirical distribution
    kde = gaussian_kde(eigenvalues)
    empirical_density = kde(x)

    # Find intersection point
    diff = empirical_density - mp_density
    cutoff_idx = np.where(diff > 0)[0]

    # If no cutoff is found, fallback to a percentile method (e.g., 90th percentile)
    if len(cutoff_idx) == 0:
        print("No eigenvalues are smaller than lambda_plus, using 90th percentile as fallback.")
        return np.percentile(eigenvalues, 90)

    return x[cutoff_idx[0]]

def denoise_correlation_matrix(returns):
    """
    Implement the de-noising algorithm using eigenvalue adjustment.
    """
    # Calculate sample covariance matrix
    cov_matrix = returns.cov().values
    correlation_matrix = returns.corr().values

    # Perform eigendecomposition
    eigenvalues, eigenvectors = np.linalg.eigh(correlation_matrix)
    eigenvalues = eigenvalues[::-1]
    eigenvectors = eigenvectors[:, ::-1]

    # Find cutoff eigenvalue
    N = len(returns.columns)
    T = len(returns)
    lambda_plus = find_eigenvalue_cutoff(eigenvalues, N, T)

    # Check if any eigenvalues are smaller than lambda_plus
    noise_indices = np.where(eigenvalues <= lambda_plus)[0]

    if len(noise_indices) == 0:
        print("No eigenvalues found smaller than or equal to lambda_plus. Using the smallest eigenvalue as fallback.")
        i = len(eigenvalues) - 1  # Use the smallest eigenvalue
    else:
        i = noise_indices[0]

    # Adjust noisy eigenvalues
    adjusted_eigenvalues = eigenvalues.copy()
    for j in range(i, N):
        adjusted_eigenvalues[j] = np.mean(eigenvalues[i:])

    # Reconstruct correlation matrix
    Lambda_tilde = np.diag(adjusted_eigenvalues)
    C_tilde = eigenvectors @ Lambda_tilde @ eigenvectors.T

    # Rescale to ensure diagonal entries are 1
    scale = np.sqrt(np.diag(C_tilde))
    C = C_tilde / (scale[:, None] @ scale[None, :])

    # Convert back to covariance matrix
    std_dev = np.sqrt(np.diag(cov_matrix))
    denoised_cov = C * (std_dev[:, None] @ std_dev[None, :])

    return denoised_cov


def create_min_variance_portfolio(returns, prev_weights=None, lambda_cost=0.0050):
    """
    Create a minimum variance portfolio with transaction costs and de-noised covariance matrix
    """
    n = len(returns.columns)

    # Calculate de-noised covariance matrix
    cov_matrix = denoise_correlation_matrix(returns)

    # If no previous weights provided, initialize with zeros
    if prev_weights is None:
        prev_weights = np.zeros(n)

    def portfolio_objective(weights, cov_matrix, prev_weights, lambda_cost):
        variance = np.dot(weights.T, np.dot(cov_matrix, weights))
        if prev_weights is not None:
            transaction_costs = lambda_cost * np.sum(np.abs(weights - prev_weights))
        else:
            transaction_costs = 0
        return variance + transaction_costs

    constraints = [
        {'type': 'eq', 'fun': lambda x: np.sum(x) - 1}
    ]

    bounds = tuple((0, 1) for _ in range(n))
    x0 = prev_weights if prev_weights is not None else np.array([1/n] * n)

    result = minimize(
        portfolio_objective,
        x0,
        args=(cov_matrix, prev_weights, lambda_cost),
        method='SLSQP',
        bounds=bounds,
        constraints=constraints,
        options={'ftol': 1e-9, 'maxiter': 1000}
    )

    if not result.success:
        print(f"Warning: Optimization failed: {result.message}")

    return result.x

def get_stock_weightage(weekly_returns, current_date, prev_weights=None):
    data_slice = weekly_returns.loc[:current_date].iloc[-50:]

    if len(data_slice) < 50:
        raise ValueError("Not enough historical data before the target date.")

    ranked_stocks = rank_stocks(data_slice)
    top_stocks = get_top_stocks(ranked_stocks)
    top_stock_returns = data_slice[top_stocks]

    # Align previous weights with current stock selection
    if prev_weights is not None:
        aligned_prev_weights = np.zeros(len(top_stocks))
        for i, stock in enumerate(top_stocks):
            if stock in prev_weights:
                aligned_prev_weights[i] = prev_weights[stock]
        prev_weights = aligned_prev_weights

    # Get optimal weights
    weights = create_min_variance_portfolio(
        top_stock_returns,
        prev_weights,
        lambda_cost=0.0050  # 50 basis points transaction cost
    )

    # Create dictionary of stock weights
    stock_weights = dict(zip(top_stocks, weights))

    # Ensure weights sum to 1 (handle numerical precision issues)
    weight_sum = sum(stock_weights.values())
    if abs(weight_sum - 1.0) > 1e-5:
        for stock in stock_weights:
            stock_weights[stock] /= weight_sum

    # Debug information
    weight_values = list(stock_weights.values())
    print(f"\nDate: {current_date}")
    print(f"Weight standard deviation: {np.std(weight_values):.4f}")
    print(f"Max weight: {max(weight_values):.4f}")
    print(f"Min weight: {min(weight_values):.4f}")

    return stock_weights

def calculate_drawdown(cumulative_returns):
    drawdown = cumulative_returns / cumulative_returns.cummax() - 1
    max_drawdown = drawdown.min()
    return drawdown, max_drawdown

def sortino_ratio(returns, target=0, periods_per_year=52):
    downside_returns = returns[returns < target]
    downside_deviation = np.sqrt((downside_returns ** 2).mean()) * np.sqrt(periods_per_year)
    annualized_return = returns.mean() * periods_per_year
    return annualized_return / downside_deviation if downside_deviation != 0 else np.nan

def calmar_ratio(annualized_return, max_drawdown):
    return annualized_return / abs(max_drawdown) if max_drawdown != 0 else np.nan

def backtest_strategy(df):
    weekly_returns = calculate_weekly_returns(df)
    portfolio_returns = []
    weightage_history = []
    prev_weights = None

    for i in range(50, len(weekly_returns)):
        current_date = weekly_returns.index[i-1]
        next_week = weekly_returns.index[i]

        try:
            weightage = get_stock_weightage(weekly_returns.iloc[:i], current_date, prev_weights)
            top_stocks = list(weightage.keys())
            weights = list(weightage.values())

            # Calculate transaction costs
            if prev_weights is not None:
                transaction_costs = 0.0050 * sum(  # 50 basis points
                    abs(weightage.get(stock, 0) - prev_weights.get(stock, 0))
                    for stock in set(weightage.keys()) | set(prev_weights.keys())
                )
            else:
                transaction_costs = 0

            # Calculate returns after transaction costs
            next_week_return = weekly_returns.loc[next_week][top_stocks].dot(weights) - transaction_costs
            portfolio_returns.append(next_week_return)
            weightage_history.append((next_week, weightage))

            prev_weights = weightage

        except ValueError as e:
            print(f"Skipping week {next_week} due to error: {e}")

    valid_index = weekly_returns.index[50:50+len(portfolio_returns)]
    return pd.Series(portfolio_returns, index=valid_index), weightage_history

def analyze_and_plot_results(portfolio_returns):
    cumulative_returns = (1 + portfolio_returns).cumprod()
    annualized_return = portfolio_returns.mean() * 52
    drawdown, max_drawdown = calculate_drawdown(cumulative_returns)

    sortino = sortino_ratio(portfolio_returns)
    calmar = calmar_ratio(annualized_return, max_drawdown)

    print(f"Annualized Return: {annualized_return:.4f}")
    print(f"Max Drawdown: {max_drawdown:.4f}")
    print(f"Sortino Ratio: {sortino:.4f}")
    print(f"Calmar Ratio: {calmar:.4f}")

    plt.figure(figsize=(10, 6))
    plt.plot(cumulative_returns, label='Cumulative Returns')
    plt.plot(drawdown, label='Drawdown', linestyle='--', color='red')
    plt.legend(loc='best')
    plt.title("Backtest Results")
    plt.show()

# Load data
df = pd.read_csv('total_df.csv', index_col='Date', parse_dates=True, dayfirst=True)

# Run backtest
portfolio_returns, weightage_history = backtest_strategy(df)

# Analyze results
analyze_and_plot_results(portfolio_returns)
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize
from scipy.cluster.hierarchy import linkage, fcluster
from scipy.spatial.distance import squareform

def correlation_distance(correlation_matrix):
    """Convert correlation matrix to distance matrix"""
    distance = np.sqrt(2 * (1 - correlation_matrix))
    return distance

def cluster_correlation_matrix(correlation_matrix, n_clusters):
    """Cluster correlation matrix using hierarchical clustering"""
    # Convert correlation to distance
    dist = correlation_distance(correlation_matrix)

    # Perform hierarchical clustering
    condensed_dist = squareform(dist)
    z = linkage(condensed_dist, method='ward')
    clusters = fcluster(z, n_clusters, criterion='maxclust')

    return clusters

def get_intra_cluster_weights(returns, clusters, lambda_cost=0.005):
    """Optimize weights within each cluster"""
    n_clusters = len(np.unique(clusters))
    n_assets = len(returns.columns)

    # Initialize the full weight matrix
    Omega_intra = np.zeros((n_clusters, n_assets))

    for k in range(n_clusters):
        # Get assets in current cluster
        cluster_mask = clusters == k + 1
        cluster_assets = returns.columns[cluster_mask]

        if len(cluster_assets) == 0:
            continue

        # Get returns for cluster assets
        cluster_returns = returns[cluster_assets]

        # Calculate covariance matrix for cluster
        cluster_cov = denoise_correlation_matrix(cluster_returns)

        # Optimize within cluster
        def cluster_objective(w):
            return np.dot(w.T, np.dot(cluster_cov, w))

        constraints = [
            {'type': 'eq', 'fun': lambda x: np.sum(x) - 1}
        ]

        bounds = tuple((0, 1) for _ in range(len(cluster_assets)))

        result = minimize(
            cluster_objective,
            x0=np.array([1/len(cluster_assets)] * len(cluster_assets)),
            method='SLSQP',
            bounds=bounds,
            constraints=constraints
        )

        # Store optimized weights in full matrix
        Omega_intra[k, cluster_mask] = result.x

    return Omega_intra

def get_reduced_covariance(returns, Omega_intra):
    """Calculate reduced covariance matrix using intra-cluster weights"""
    cov = denoise_correlation_matrix(returns)
    V_reduced = Omega_intra @ cov @ Omega_intra.T
    return V_reduced

def get_inter_cluster_weights(V_reduced, lambda_cost=0.005):
    """Optimize weights between clusters"""
    n_clusters = len(V_reduced)

    def cluster_objective(w):
        return np.dot(w.T, np.dot(V_reduced, w))

    constraints = [
        {'type': 'eq', 'fun': lambda x: np.sum(x) - 1}
    ]

    bounds = tuple((0, 1) for _ in range(n_clusters))

    result = minimize(
        cluster_objective,
        x0=np.array([1/n_clusters] * n_clusters),
        method='SLSQP',
        bounds=bounds,
        constraints=constraints
    )

    return result.x

def nested_portfolio_optimization(returns, n_clusters=5, lambda_cost=0.005):
    """Implement nested clustered optimization"""
    # Get correlation matrix
    correlation_matrix = returns.corr()

    # Step 2: Cluster correlation matrix
    clusters = cluster_correlation_matrix(correlation_matrix, n_clusters)

    # Step 3: Intra-cluster optimization
    Omega_intra = get_intra_cluster_weights(returns, clusters, lambda_cost)

    # Step 4: Get reduced covariance matrix
    V_reduced = get_reduced_covariance(returns, Omega_intra)

    # Step 5: Inter-cluster optimization
    Omega_inter = get_inter_cluster_weights(V_reduced, lambda_cost)

    # Step 6: Final weight allocation
    final_weights = Omega_intra.T @ Omega_inter

    # Create dictionary of weights
    weight_dict = dict(zip(returns.columns, final_weights))

    return weight_dict

# Modify the existing get_stock_weightage function to use nested optimization
def get_stock_weightage(weekly_returns, current_date, prev_weights=None):
    data_slice = weekly_returns.loc[:current_date].iloc[-50:]

    if len(data_slice) < 50:
        raise ValueError("Not enough historical data before the target date.")

    ranked_stocks = rank_stocks(data_slice)
    top_stocks = get_top_stocks(ranked_stocks)
    top_stock_returns = data_slice[top_stocks]

    # Use nested optimization instead of simple min variance
    stock_weights = nested_portfolio_optimization(
        top_stock_returns,
        n_clusters=5,  # Can be adjusted based on portfolio size
        lambda_cost=0.0050
    )

    # Ensure weights sum to 1
    weight_sum = sum(stock_weights.values())
    if abs(weight_sum - 1.0) > 1e-5:
        for stock in stock_weights:
            stock_weights[stock] /= weight_sum

    # Debug information
    weight_values = list(stock_weights.values())
    print(f"\nDate: {current_date}")
    print(f"Weight standard deviation: {np.std(weight_values):.4f}")
    print(f"Max weight: {max(weight_values):.4f}")
    print(f"Min weight: {min(weight_values):.4f}")

    return stock_weights
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from scipy.cluster.hierarchy import linkage, fcluster
from scipy.spatial.distance import squareform

def calculate_weekly_returns(df):
    weekly_data = df.resample('W').last()
    weekly_returns = weekly_data.pct_change().dropna()
    return weekly_returns

def rank_stocks(weekly_returns, window=50):
    avg_returns = weekly_returns.rolling(window=window).mean()
    latest_avg_returns = avg_returns.iloc[-1].sort_values(ascending=False)
    return latest_avg_returns.index.tolist()

def get_top_stocks(ranked_stocks, n=20):
    return ranked_stocks[:n]

def marcenko_pastur_pdf(x, Q, sigma=1):
    y = 1/Q
    b = np.power(sigma * (1 + np.sqrt(1/Q)), 2)
    a = np.power(sigma * (1 - np.sqrt(1/Q)), 2)
    return (Q/(2*np.pi*sigma*sigma*x)) * np.sqrt((b-x)*(x-a)) if a <= x <= b else 0

def find_eigenvalue_cutoff(eigenvalues, N, T):
    Q = N/T
    sigma = np.std(eigenvalues)
    x = np.linspace(0, max(eigenvalues), 1000)
    mp_density = [marcenko_pastur_pdf(xi, Q, sigma) for xi in x]
    kde = gaussian_kde(eigenvalues)
    empirical_density = kde(x)
    diff = empirical_density - mp_density
    cutoff_idx = np.where(diff > 0)[0]

    if len(cutoff_idx) == 0:
        return np.percentile(eigenvalues, 90)

    return x[cutoff_idx[0]]

def denoise_correlation_matrix(returns):
    cov_matrix = returns.cov().values
    correlation_matrix = returns.corr().values

    eigenvalues, eigenvectors = np.linalg.eigh(correlation_matrix)
    eigenvalues = eigenvalues[::-1]
    eigenvectors = eigenvectors[:, ::-1]

    N = len(returns.columns)
    T = len(returns)
    lambda_plus = find_eigenvalue_cutoff(eigenvalues, N, T)

    noise_indices = np.where(eigenvalues <= lambda_plus)[0]

    if len(noise_indices) == 0:
        i = len(eigenvalues) - 1
    else:
        i = noise_indices[0]

    adjusted_eigenvalues = eigenvalues.copy()
    for j in range(i, N):
        adjusted_eigenvalues[j] = np.mean(eigenvalues[i:])

    Lambda_tilde = np.diag(adjusted_eigenvalues)
    C_tilde = eigenvectors @ Lambda_tilde @ eigenvectors.T

    scale = np.sqrt(np.diag(C_tilde))
    C = C_tilde / (scale[:, None] @ scale[None, :])

    std_dev = np.sqrt(np.diag(cov_matrix))
    denoised_cov = C * (std_dev[:, None] @ std_dev[None, :])

    return denoised_cov

def correlation_distance(correlation_matrix):
    distance = np.sqrt(2 * (1 - correlation_matrix))
    return distance

def cluster_correlation_matrix(correlation_matrix, n_clusters):
    dist = correlation_distance(correlation_matrix)
    condensed_dist = squareform(dist)
    z = linkage(condensed_dist, method='ward')
    clusters = fcluster(z, n_clusters, criterion='maxclust')
    return clusters

def get_intra_cluster_weights(returns, clusters, lambda_cost=0.005):
    n_clusters = len(np.unique(clusters))
    n_assets = len(returns.columns)
    Omega_intra = np.zeros((n_clusters, n_assets))

    for k in range(n_clusters):
        cluster_mask = clusters == k + 1
        cluster_assets = returns.columns[cluster_mask]

        if len(cluster_assets) == 0:
            continue

        cluster_returns = returns[cluster_assets]
        cluster_cov = denoise_correlation_matrix(cluster_returns)

        def cluster_objective(w):
            return np.dot(w.T, np.dot(cluster_cov, w))

        constraints = [{'type': 'eq', 'fun': lambda x: np.sum(x) - 1}]
        bounds = tuple((0, 1) for _ in range(len(cluster_assets)))

        result = minimize(
            cluster_objective,
            x0=np.array([1/len(cluster_assets)] * len(cluster_assets)),
            method='SLSQP',
            bounds=bounds,
            constraints=constraints
        )

        Omega_intra[k, cluster_mask] = result.x

    return Omega_intra

def get_reduced_covariance(returns, Omega_intra):
    cov = denoise_correlation_matrix(returns)
    V_reduced = Omega_intra @ cov @ Omega_intra.T
    return V_reduced

def get_inter_cluster_weights(V_reduced, lambda_cost=0.005):
    n_clusters = len(V_reduced)

    def cluster_objective(w):
        return np.dot(w.T, np.dot(V_reduced, w))

    constraints = [{'type': 'eq', 'fun': lambda x: np.sum(x) - 1}]
    bounds = tuple((0, 1) for _ in range(n_clusters))

    result = minimize(
        cluster_objective,
        x0=np.array([1/n_clusters] * n_clusters),
        method='SLSQP',
        bounds=bounds,
        constraints=constraints
    )

    return result.x

def nested_portfolio_optimization(returns, n_clusters=5, lambda_cost=0.005):
    correlation_matrix = returns.corr()
    clusters = cluster_correlation_matrix(correlation_matrix, n_clusters)
    Omega_intra = get_intra_cluster_weights(returns, clusters, lambda_cost)
    V_reduced = get_reduced_covariance(returns, Omega_intra)
    Omega_inter = get_inter_cluster_weights(V_reduced, lambda_cost)
    final_weights = Omega_intra.T @ Omega_inter
    weight_dict = dict(zip(returns.columns, final_weights))
    return weight_dict

def get_stock_weightage(weekly_returns, current_date, prev_weights=None):
    data_slice = weekly_returns.loc[:current_date].iloc[-50:]

    if len(data_slice) < 50:
        raise ValueError("Not enough historical data before the target date.")

    ranked_stocks = rank_stocks(data_slice)
    top_stocks = get_top_stocks(ranked_stocks)
    top_stock_returns = data_slice[top_stocks]

    stock_weights = nested_portfolio_optimization(
        top_stock_returns,
        n_clusters=5,
        lambda_cost=0.0050
    )

    weight_sum = sum(stock_weights.values())
    if abs(weight_sum - 1.0) > 1e-5:
        for stock in stock_weights:
            stock_weights[stock] /= weight_sum

    weight_values = list(stock_weights.values())
    print(f"\nDate: {current_date}")
    print(f"Weight standard deviation: {np.std(weight_values):.4f}")
    print(f"Max weight: {max(weight_values):.4f}")
    print(f"Min weight: {min(weight_values):.4f}")

    return stock_weights

def calculate_drawdown(cumulative_returns):
    drawdown = cumulative_returns / cumulative_returns.cummax() - 1
    max_drawdown = drawdown.min()
    return drawdown, max_drawdown

def sortino_ratio(returns, target=0, periods_per_year=52):
    downside_returns = returns[returns < target]
    downside_deviation = np.sqrt((downside_returns ** 2).mean()) * np.sqrt(periods_per_year)
    annualized_return = returns.mean() * periods_per_year
    return annualized_return / downside_deviation if downside_deviation != 0 else np.nan

def calmar_ratio(annualized_return, max_drawdown):
    return annualized_return / abs(max_drawdown) if max_drawdown != 0 else np.nan

def backtest_strategy(df):
    weekly_returns = calculate_weekly_returns(df)
    portfolio_returns = []
    weightage_history = []
    prev_weights = None

    for i in range(50, len(weekly_returns)):
        current_date = weekly_returns.index[i-1]
        next_week = weekly_returns.index[i]

        try:
            weightage = get_stock_weightage(weekly_returns.iloc[:i], current_date, prev_weights)
            top_stocks = list(weightage.keys())
            weights = list(weightage.values())

            if prev_weights is not None:
                transaction_costs = 0.0050 * sum(
                    abs(weightage.get(stock, 0) - prev_weights.get(stock, 0))
                    for stock in set(weightage.keys()) | set(prev_weights.keys())
                )
            else:
                transaction_costs = 0

            next_week_return = weekly_returns.loc[next_week][top_stocks].dot(weights) - transaction_costs
            portfolio_returns.append(next_week_return)
            weightage_history.append((next_week, weightage))

            prev_weights = weightage

        except ValueError as e:
            print(f"Skipping week {next_week} due to error: {e}")

    valid_index = weekly_returns.index[50:50+len(portfolio_returns)]
    return pd.Series(portfolio_returns, index=valid_index), weightage_history

def analyze_and_plot_results(portfolio_returns):
    cumulative_returns = (1 + portfolio_returns).cumprod()
    annualized_return = portfolio_returns.mean() * 52
    drawdown, max_drawdown = calculate_drawdown(cumulative_returns)

    sortino = sortino_ratio(portfolio_returns)
    calmar = calmar_ratio(annualized_return, max_drawdown)

    print(f"Annualized Return: {annualized_return:.4f}")
    print(f"Max Drawdown: {max_drawdown:.4f}")
    print(f"Sortino Ratio: {sortino:.4f}")
    print(f"Calmar Ratio: {calmar:.4f}")

    plt.figure(figsize=(10, 6))
    plt.plot(cumulative_returns, label='Cumulative Returns')
    plt.plot(drawdown, label='Drawdown', linestyle='--', color='red')
    plt.legend(loc='best')
    plt.title("Backtest Results")
    plt.show()

# Keep all the previous code as is, just replace the main execution block at the end with:

# Main execution
if __name__ == "__main__":
    # Load data
    df = pd.read_csv('total_df.csv', index_col='Date', parse_dates=True, dayfirst=True)

    # Run backtest
    portfolio_returns, weightage_history = backtest_strategy(df)

    # Analyze results
    analyze_and_plot_results(portfolio_returns)
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from scipy.cluster.hierarchy import linkage, fcluster
from scipy.spatial.distance import squareform

def calculate_weekly_returns(df):
    weekly_data = df.resample('W').last()
    weekly_returns = weekly_data.pct_change().dropna()
    return weekly_returns

def rank_stocks(weekly_returns, window=50):
    avg_returns = weekly_returns.rolling(window=window).mean()
    latest_avg_returns = avg_returns.iloc[-1].sort_values(ascending=False)
    return latest_avg_returns.index.tolist()

def get_top_stocks(ranked_stocks, n=20):
    return ranked_stocks[:n]

def marcenko_pastur_pdf(x, Q, sigma=1):
    """Calculate Marcenko-Pastur probability density at x with error handling"""
    if x <= 0:
        return 0
    y = 1/Q
    b = np.power(sigma * (1 + np.sqrt(1/Q)), 2)  # Upper bound
    a = np.power(sigma * (1 - np.sqrt(1/Q)), 2)  # Lower bound

    try:
        if a <= x <= b:
            return (Q/(2*np.pi*sigma*sigma*x)) * np.sqrt((b-x)*(x-a))
        return 0
    except:
        return 0

def find_eigenvalue_cutoff(eigenvalues, N, T):
    """Find the eigenvalue cutoff point with improved robustness"""
    if len(eigenvalues) < 2:
        return max(eigenvalues) if len(eigenvalues) > 0 else 1.0

    Q = N/T
    sigma = np.std(eigenvalues)

    try:
        x = np.linspace(max(1e-10, min(eigenvalues)), max(eigenvalues), 1000)
        mp_density = np.array([marcenko_pastur_pdf(xi, Q, sigma) for xi in x])

        if len(eigenvalues) > 1:
            kde = gaussian_kde(eigenvalues)
            empirical_density = kde(x)
        else:
            return eigenvalues[0]

        diff = empirical_density - mp_density
        cutoff_idx = np.where(diff > 0)[0]

        if len(cutoff_idx) == 0:
            return np.percentile(eigenvalues, 90)

        return x[cutoff_idx[0]]
    except:
        return np.percentile(eigenvalues, 90)

def denoise_correlation_matrix(returns):
    """Denoise the correlation matrix using eigenvalue adjustment"""
    if len(returns.columns) < 2:
        return returns.cov().values

    cov_matrix = returns.cov().values
    correlation_matrix = returns.corr().values

    eigenvalues, eigenvectors = np.linalg.eigh(correlation_matrix)
    eigenvalues = eigenvalues[::-1]
    eigenvectors = eigenvectors[:, ::-1]

    N = len(returns.columns)
    T = len(returns)
    lambda_plus = find_eigenvalue_cutoff(eigenvalues, N, T)

    noise_indices = np.where(eigenvalues <= lambda_plus)[0]

    if len(noise_indices) == 0:
        i = len(eigenvalues) - 1
    else:
        i = noise_indices[0]

    adjusted_eigenvalues = eigenvalues.copy()
    for j in range(i, N):
        adjusted_eigenvalues[j] = np.mean(eigenvalues[i:])

    Lambda_tilde = np.diag(adjusted_eigenvalues)
    C_tilde = eigenvectors @ Lambda_tilde @ eigenvectors.T

    scale = np.sqrt(np.diag(C_tilde))
    C = C_tilde / (scale[:, None] @ scale[None, :])

    std_dev = np.sqrt(np.diag(cov_matrix))
    denoised_cov = C * (std_dev[:, None] @ std_dev[None, :])

    return denoised_cov

def correlation_distance(correlation_matrix):
    distance = np.sqrt(2 * (1 - correlation_matrix))
    return distance

def cluster_correlation_matrix(correlation_matrix, n_clusters):
    dist = correlation_distance(correlation_matrix)

    # Handle the case where we have fewer assets than clusters
    if len(correlation_matrix) <= n_clusters:
        return np.arange(1, len(correlation_matrix) + 1)

    condensed_dist = squareform(dist)
    z = linkage(condensed_dist, method='ward')
    clusters = fcluster(z, n_clusters, criterion='maxclust')
    return clusters

def get_intra_cluster_weights(returns, clusters, lambda_cost=0.005):
    n_clusters = len(np.unique(clusters))
    n_assets = len(returns.columns)
    Omega_intra = np.zeros((n_clusters, n_assets))

    for k in range(n_clusters):
        cluster_mask = clusters == k + 1
        cluster_assets = returns.columns[cluster_mask]

        if len(cluster_assets) == 0:
            continue

        cluster_returns = returns[cluster_assets]

        if len(cluster_assets) == 1:
            Omega_intra[k, cluster_mask] = 1
            continue

        cluster_cov = denoise_correlation_matrix(cluster_returns)

        def cluster_objective(w):
            return np.dot(w.T, np.dot(cluster_cov, w))

        constraints = [{'type': 'eq', 'fun': lambda x: np.sum(x) - 1}]
        bounds = tuple((0, 1) for _ in range(len(cluster_assets)))

        try:
            result = minimize(
                cluster_objective,
                x0=np.array([1/len(cluster_assets)] * len(cluster_assets)),
                method='SLSQP',
                bounds=bounds,
                constraints=constraints
            )

            if result.success:
                Omega_intra[k, cluster_mask] = result.x
            else:
                Omega_intra[k, cluster_mask] = 1/len(cluster_assets)
        except:
            Omega_intra[k, cluster_mask] = 1/len(cluster_assets)

    return Omega_intra

def get_reduced_covariance(returns, Omega_intra):
    cov = denoise_correlation_matrix(returns)
    V_reduced = Omega_intra @ cov @ Omega_intra.T
    return V_reduced

def get_inter_cluster_weights(V_reduced, lambda_cost=0.005):
    n_clusters = len(V_reduced)

    if n_clusters == 1:
        return np.array([1.0])

    def cluster_objective(w):
        return np.dot(w.T, np.dot(V_reduced, w))

    constraints = [{'type': 'eq', 'fun': lambda x: np.sum(x) - 1}]
    bounds = tuple((0, 1) for _ in range(n_clusters))

    try:
        result = minimize(
            cluster_objective,
            x0=np.array([1/n_clusters] * n_clusters),
            method='SLSQP',
            bounds=bounds,
            constraints=constraints
        )

        if result.success:
            return result.x
        else:
            return np.array([1/n_clusters] * n_clusters)
    except:
        return np.array([1/n_clusters] * n_clusters)

def nested_portfolio_optimization(returns, n_clusters=5, lambda_cost=0.005):
    """Nested portfolio optimization with improved robustness"""
    try:
        if len(returns.columns) < n_clusters:
            n_clusters = max(2, len(returns.columns) // 2)

        correlation_matrix = returns.corr()
        clusters = cluster_correlation_matrix(correlation_matrix, n_clusters)
        Omega_intra = get_intra_cluster_weights(returns, clusters, lambda_cost)

        if np.all(Omega_intra == 0):
            # Fallback to equal weights if clustering fails
            equal_weights = {col: 1.0/len(returns.columns) for col in returns.columns}
            return equal_weights

        V_reduced = get_reduced_covariance(returns, Omega_intra)
        Omega_inter = get_inter_cluster_weights(V_reduced, lambda_cost)
        final_weights = Omega_intra.T @ Omega_inter
        weight_dict = dict(zip(returns.columns, final_weights))

        return weight_dict
    except Exception as e:
        print(f"Error in nested optimization: {str(e)}")
        # Fallback to equal weights
        equal_weights = {col: 1.0/len(returns.columns) for col in returns.columns}
        return equal_weights

def get_stock_weightage(weekly_returns, current_date, prev_weights=None, min_stocks=5):
    """Get stock weightage with improved handling of limited data"""
    try:
        data_slice = weekly_returns.loc[:current_date].iloc[-50:]

        if len(data_slice) < 10:  # Require at least 10 weeks of data
            raise ValueError("Not enough historical data")

        ranked_stocks = rank_stocks(data_slice)
        top_stocks = get_top_stocks(ranked_stocks)

        if len(top_stocks) < min_stocks:
            raise ValueError(f"Not enough stocks (minimum {min_stocks} required)")

        top_stock_returns = data_slice[top_stocks]

        # Determine number of clusters based on number of stocks
        n_clusters = min(5, len(top_stocks) // 2)

        stock_weights = nested_portfolio_optimization(
            top_stock_returns,
            n_clusters=max(2, n_clusters),  # At least 2 clusters
            lambda_cost=0.0050
        )

        weight_sum = sum(stock_weights.values())
        if abs(weight_sum - 1.0) > 1e-5:
            for stock in stock_weights:
                stock_weights[stock] /= weight_sum

        weight_values = list(stock_weights.values())
        print(f"\nDate: {current_date}")
        print(f"Number of stocks: {len(stock_weights)}")
        print(f"Weight standard deviation: {np.std(weight_values):.4f}")
        print(f"Max weight: {max(weight_values):.4f}")
        print(f"Min weight: {min(weight_values):.4f}")

        return stock_weights

    except Exception as e:
        print(f"Error in get_stock_weightage for date {current_date}: {str(e)}")
        if prev_weights is not None:
            print("Using previous weights")
            return prev_weights
        else:
            # Return equal weights for top stocks as fallback
            try:
                ranked_stocks = rank_stocks(weekly_returns.loc[:current_date].iloc[-10:])
                fallback_stocks = get_top_stocks(ranked_stocks, n=min_stocks)
                equal_weights = {stock: 1.0/len(fallback_stocks) for stock in fallback_stocks}
                return equal_weights
            except:
                print("Fallback strategy failed")
                return {}

def calculate_drawdown(cumulative_returns):
    rolling_max = cumulative_returns.cummax()
    drawdown = cumulative_returns / rolling_max - 1
    max_drawdown = drawdown.min()
    return drawdown, max_drawdown

def sortino_ratio(returns, target=0, periods_per_year=52):
    downside_returns = returns[returns < target]
    if len(downside_returns) == 0:
        return np.inf if returns.mean() > 0 else -np.inf
    downside_deviation = np.sqrt((downside_returns ** 2).mean()) * np.sqrt(periods_per_year)
    excess_return = (returns.mean() - target) * periods_per_year
    return excess_return / downside_deviation if downside_deviation != 0 else np.inf

def calmar_ratio(annualized_return, max_drawdown):
    return -annualized_return / max_drawdown if max_drawdown != 0 else np.inf

def backtest_strategy(df):
    """Backtest strategy with improved error handling"""
    weekly_returns = calculate_weekly_returns(df)
    portfolio_returns = []
    weightage_history = []
    prev_weights = None

    for i in range(50, len(weekly_returns)):
        current_date = weekly_returns.index[i-1]
        next_week = weekly_returns.index[i]

        try:
            weightage = get_stock_weightage(weekly_returns.iloc[:i], current_date, prev_weights)

            if not weightage:  # Skip if weightage is empty
                print(f"Skipping week {next_week} due to empty weightage")
                continue

            top_stocks = list(weightage.keys())
            weights = list(weightage.values())

            if prev_weights is not None:
                transaction_costs = 0.0050 * sum(
                    abs(weightage.get(stock, 0) - prev_weights.get(stock, 0))
                    for stock in set(weightage.keys()) | set(prev_weights.keys())
                )
            else:
                transaction_costs = 0

            next_week_return = weekly_returns.loc[next_week][top_stocks].dot(weights) - transaction_costs
            portfolio_returns.append(next_week_return)
            weightage_history.append((next_week, weightage))

            prev_weights = weightage

        except Exception as e:
            print(f"Skipping week {next_week} due to error: {str(e)}")

    if not portfolio_returns:
        raise ValueError("No valid portfolio returns generated during backtest")

    valid_index = weekly_returns.index[50:50+len(portfolio_returns)]
    return pd.Series(portfolio_returns, index=valid_index), weightage_history

import pandas as pd
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from scipy.cluster.hierarchy import linkage, fcluster
from scipy.spatial.distance import squareform

# ... [All previously defined functions remain the same] ...

def analyze_and_plot_results(portfolio_returns):
    cumulative_returns = (1 + portfolio_returns).cumprod()
    annualized_return = portfolio_returns.mean() * 52  # 52 weeks in a year
    drawdown, max_drawdown = calculate_drawdown(cumulative_returns)

    sortino = sortino_ratio(portfolio_returns)
    calmar = calmar_ratio(annualized_return, max_drawdown)

    print(f"Annualized Return: {annualized_return:.4f}")
    print(f"Max Drawdown: {max_drawdown:.4f}")
    print(f"Sortino Ratio: {sortino:.4f}")
    print(f"Calmar Ratio: {calmar:.4f}")

    plt.figure(figsize=(15, 10))

    # Plot cumulative returns
    plt.subplot(2, 1, 1)
    plt.plot(cumulative_returns.index, cumulative_returns.values)
    plt.title('Cumulative Returns')
    plt.xlabel('Date')
    plt.ylabel('Cumulative Returns')

    # Plot drawdown
    plt.subplot(2, 1, 2)
    plt.plot(drawdown.index, drawdown.values)
    plt.title('Drawdown')
    plt.xlabel('Date')
    plt.ylabel('Drawdown')

    plt.tight_layout()
    plt.show()

def main():
    # Load your data here. For example:
    # df = pd.read_csv('your_stock_data.csv', index_col='Date', parse_dates=True)

    # Perform backtesting
    portfolio_returns, weightage_history = backtest_strategy(df)

    # Analyze and plot results
    analyze_and_plot_results(portfolio_returns)

    # You can add more analysis or plotting of weightage_history here if needed

if __name__ == "__main__":
    main()
```
